```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers = 4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape
```

```python
data[0][0:10] # see the actual data
```

```python
data[1][0:10]  # tells you which group point belongs to
```

```python
plt.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_ 
```

```python
kmeans.labels_   # labels it believes to be true for the clusters
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_ , cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1] , cmap='rainbow')
```

```python
# note that colors are just local to each plot, so they don't mean anything
```

```python
# repeat with kmeans = 2
kmeans = KMeans(n_clusters=2)
kmeans.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_ , cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1] , cmap='rainbow')
```

```python
# repeat with kmeans = 3
kmeans = KMeans(n_clusters=3)
kmeans.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_ , cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1] , cmap='rainbow')
```
