```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# FUNCTIONAL
plt.plot(x,y)
plt.xlabel('X Label')
plt.title('Title')
plt.show() #only need to use if not using Jupyter notebook
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented method
fig = plt.figure()  # created blank canvas object

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes1.set_title('LARGER PLOT')
axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)  # positions subplots automatically

#axes.plot(x,y)
plt.tight_layout()  # fixes overlaps
for current_ax in axes:
    current_ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)  # positions subplots automatically

axes[0].plot(x,y)
axes[1].plot(y,x)

    
axes # just a list of axes object, so can iterate through it
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8,2))

axes[0].plot(x,y)

axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png', dpi = 200)
```

```python
fig = plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label='X Squared')
ax.plot(x,x**3, label='X Cubed')

ax.legend(loc=(0.1,0.1))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y, color = 'purple', linewidth=3
 , alpha=0.5,
       linestyle='--',
       marker='o',
       markersize=10,
       markerfacecolor = 'yellow',
       markeredgewidth=3,
       markeredgecolor = 'green')  #alpha adds transperency

ax.set_xlim([0,3])
ax.set_ylim([0,12])
```
